**Import Libraries and Load Data:**

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
#Loading dataset
data = pd.read_csv('Airbnb NYC 2019.csv')
```

```python

# Initial Data Overview
print("First few rows of the dataset:")
display(data.head())  # Display the first few rows to understand the structure of the data
```

```python
print("\nLast few rows of the dataset:")
display(data.tail())  # Display the last few rows to check data consistency
```

```python
print("\nColumn Names:")
print(data.columns)  # Print column names to get an overview of the dataset features
```

**Data Description:**
    
+ **id:**Unique ID for the listing.
+ **name:** Name of the listing.
+ **hostid:** Unique ID for the host.
+ **host_name:** Name of the host.
+ **neighbourhood_group:** Location of the listing (city/region).
+ **neighbourhood:** Specific area or neighborhood within the location.
+ **latitude:** Latitude coordinate of the listing.
+ **longitude:** Longitude coordinate of the listing.
+ **room_type:**Type of room or listing (e.g., entire home, private room).
+ **price:**Price of the listing.
+ **minimum_nights:** Minimum number of nights required for booking.
+ **number_of_reviews:**Total number of reviews for the listing.
+ **last_review:**Date of the most recent review.
+ **reviews_per_month:** Average number of reviews per month.
+ **calculated_host_listings_count:** Total number of listings the host has.
+ **availability_365:**Number of days the listing is available in a year.

```python

# Copying data for processing
df = data.copy()  # Create a copy of the dataset to work with
```

```python

# Basic Info and Descriptive Statistics
print("\nData Info:")
df.info()  # Display data types and non-null counts for each column
```

```python
df.describe()
```

```python
print("\nDescriptive Statistics:")
display(df.describe().T.style.background_gradient())  # Generate summary statistics for numerical columns
```

**Data Cleaning**

```python
# Checking for Duplicates
print("\nNumber of duplicate rows:", df.duplicated().sum())  # Check for duplicate rows in the dataset
```

```python
# Shape of the DataFrame
print("\nShape of the data:", df.shape)  # Output the dataset's dimensions
```

```python
# Checking Missing Values
print("\nMissing Values:")
print(df.isnull().sum())  # Count missing values in each column
```

```python

# Data Cleaning: Handling Missing Values
df['name'].replace(np.nan, 'other hotel', inplace=True)
df['host_name'].replace(np.nan, 'other', inplace=True)
df['last_review'].replace(np.nan, 'not reviewed', inplace=True)
df['reviews_per_month'].replace(np.nan, 0, inplace=True)
```

**NO NAN VALUES**

```python
df.info()
```

**We dont need id, name and last_reviewed as it is completely unique and has no significance for visualization**

```python
# Drop columns if they exist
columns_to_drop = ['id', 'last_review','name']
df.drop([col for col in columns_to_drop if col in df.columns], axis=1)
```

```python
print(df.columns)  # Display columns after removing unnecessary ones
```

```python
df.isnull().sum()
```

**Data Visualizations**

**1. Distribution of Room Types:**

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='room_type', palette='Set2')
plt.title('Distribution of Room Types')
plt.show()  # Visualize the number of listings by each room type
```

Remark: This plot reveals the distribution of room types in the dataset. It is evident that Entire home/apt is the most common room type, suggesting that a significant number of Airbnb guests prefer to rent out entire apartments or houses.


```python
df['price'].describe()
```

**2. Price Distribution:**

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['price'], bins=50, kde=True, color='skyblue')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.show()  # Show the distribution of prices for listings
```

Remark: The price distribution is highly skewed, with most listings priced at lower values, but there are a few high-priced listings that create a long tail. This indicates a wide range of prices across different types of listings, with a concentration of cheaper listings



**3.Price Distribution by Room Type:**

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='room_type', y='price', palette='coolwarm')
plt.title('Price Distribution by Room Type')
plt.ylim(0, 1000)  # Adjust y-axis to remove outliers if needed
plt.show()  # Compare price distributions across room types
```

Remark: The boxplot shows that Entire home/apt has a higher median price compared to other room types like Private room and Shared room. The presence of outliers in the Entire home/apt category suggests that luxury listings might skew the price distribution.



**4. Neighborhood Preferences:**

```python

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='neighbourhood_group', palette='Set3')
plt.title('Distribution of Listings by Neighborhood Group')
plt.show()  # Visualize the count of listings across neighborhood groups
```

Remark: This count plot indicates that the highest number of Airbnb listings are located in Manhattan and Brooklyn, followed by Queens, while Staten Island and The Bronx have relatively fewer listings. This could be indicative of demand being higher in more central locations.



**Get Unique Neighborhoods:**

```python
neighborhoods = data['neighbourhood_group'].unique()
print(neighborhoods)
```

**5. Room Type Distribution by Neighborhood Group:**

```python
# 5. Room Type Distribution by Neighborhood Group
neighborhoods = df['neighbourhood_group'].unique()
for neighborhood in neighborhoods:
    subset = df[df['neighbourhood_group'] == neighborhood]
    plt.figure(figsize=(6, 4))
    sns.countplot(data=subset, x='room_type', palette='Set3')
    plt.title(f'Room Type Distribution in {neighborhood}')
    plt.show()  # Show room type distribution within each neighborhood group
```

Remark: These visualizations display how room types are distributed across different neighborhoods. For example, Entire home/apt is most popular in central neighborhoods like Manhattan, while Private rooms tend to dominate in neighborhoods like The Bronx and Staten Island.

**Exploring Assumptions:**

** Are private rooms preferred over other room types?**

```python
room_type_counts = data['room_type'].value_counts(normalize=True) * 100
print(room_type_counts)

room_type_counts = data['room_type'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Set3'))
plt.title('Room Type Distribution')
plt.show()
```

Remark: The pie chart offers a clear visualization of the relative proportions of each room type. Entire home/apt constitutes the majority of listings, confirming the trend observed earlier in the count plot.

**Room Type vs Availability Heatmap**

```python
# Create pivot table for room type and availability
room_availability = data.pivot_table(index='room_type', columns='neighbourhood_group', values='availability_365', aggfunc='mean')

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(room_availability, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Average Availability of Room Types by Neighborhood')
plt.show()
```

Remark: The heatmap shows that Entire home/apt has a higher average availability in neighborhoods like Brooklyn and Queens, while Private rooms tend to have more availability in central neighborhoods like Manhattan.



**Exploring Pricing Trends:**

```python
sns.boxplot(data=data, x='neighbourhood_group', y='price', palette='Set2')
plt.title('Price Distribution Across Neighborhoods')
plt.ylim(0, 1000)  # Adjust y-axis to focus on typical price ranges
plt.show()
```

Remark: This boxplot clearly shows that Manhattan has the highest price range, with prices often exceeding , while other boroughs like Queens and Staten Island have lower average prices, reflecting their less central locations.



**Exploring Availability:**

```python
sns.boxplot(data=data, x='room_type', y='availability_365', palette='Set3')
plt.title('Availability of Listings by Room Type')
plt.show()
```

Remark: Entire home/apt listings tend to have a lower availability compared to Private rooms, suggesting that entire apartments are in higher demand and booked for longer periods, while private rooms might be more readily available throughout the year.



**Neighborhood-Level Analysis:**

```python
# Bar Plot for Average Prices by Neighborhood
neighborhood_avg_price = data.groupby('neighbourhood_group')['price'].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=neighborhood_avg_price.index, y=neighborhood_avg_price.values, palette='coolwarm')
plt.title('Average Price by Neighborhood')
plt.ylabel('Average Price ($)')
plt.show()


```

Remark: The bar plot confirms that Manhattan has the highest average price per night, followed by Brooklyn. Other boroughs like Queens and Staten Island have more affordable listings on average.



**Correlation Analysis:**

```python
# Correlation matrix
corr_matrix = data[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365']].corr()

# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Key Variables')
plt.show()
```

Remark: The correlation matrix reveals strong relationships between variables such as price and number_of_reviews, and weaker correlations with minimum_nights and availability_365. Listings with more reviews tend to have higher prices, suggesting that more popular listings attract higher demand and higher pricing.

**Price vs. Number of Reviews Scatter Plot:**

```python
#Explore if the number of reviews impacts the price of listings. This can show if more popular listings (more reviews) tend to be more expensive or cheaper

plt.figure(figsize=(8, 5))
sns.scatterplot(data=data, x='number_of_reviews', y='price', hue='room_type', palette='bright', alpha=0.7)
plt.title('Number of Reviews vs Price by Room Type')
plt.ylim(0, 1000)  # Focus on typical price range
plt.show()
```

Remark: This scatter plot suggests that there is a positive relationship between the number of reviews and price, particularly for Entire home/apt listings. Listings with more reviews tend to have a higher price, indicating greater popularity and demand.



**Top 10 Most Reviewed Listings**

```python
# Sort by number of reviews and get top 10
top_10_reviews = data.sort_values(by='number_of_reviews', ascending=False).head(10)

# Bar Plot for top 10 reviewed listings
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_reviews['name'], y=top_10_reviews['number_of_reviews'], palette='Set2')
plt.title('Top 10 Most Reviewed Listings')
plt.xticks(rotation=90)
plt.ylabel('Number of Reviews')
plt.show()
```

Remark: This bar plot highlights the top 10 most reviewed listings, which often have higher prices and more frequent bookings. These listings are typically the most popular, suggesting they are well-regarded by guests and have a consistent track record of high demand.
